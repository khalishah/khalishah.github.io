```python
import pandas as pd 
import numpy as np

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
df = pd.read_csv('population2020.csv')
df.head()
```

```python
df['Subzone'] = df['Subzone'].str.replace('0', '-')
df['PopMale'] = df['PopMale'].replace('-', 0)
df['PopMale'] = df['PopMale'].astype(int)
df['PopFemale'] = df['PopFemale'].replace('-', 0)
df['PopFemale'] = df['PopFemale'].astype(int)
```

```python
df.sort_values(by = 'PopFemale', ascending = False)
```

```python
df.to_csv('population2020c.csv', index = False)
```
